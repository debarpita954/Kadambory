```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
cancer_data = pd.read_csv('breast_cancer_data.csv')
```

```python
cancer_data.head(20)
```

```python
sns.countplot(x='diagnosis', data=cancer_data);
```

```python
sns.heatmap(cancer_data.isnull());
```

```python
cancer_data.drop('id', axis=1, inplace=True)
```

```python
cancer_data.head(5)
```

```python
cancer_data.replace(to_replace ="M", value = 1, inplace=True)
```

```python
cancer_data.replace(to_replace ="B", value = 0, inplace = True)
```

```python
cancer_data.head(5)
```

```python
cancer_data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
X = cancer_data.drop('diagnosis', axis=1)
Y = cancer_data['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
analytics_data = LogisticRegression()
```

```python
analytics_data.fit(X_train, Y_train)
```

```python
predictions = analytics_data.predict(X_test)
print(predictions)
```
